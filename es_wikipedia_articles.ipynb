```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import wikipedia as wk
```

```python
# Getting all the main topics
base_url = 'https://es.wikipedia.org'
topics_url = 'https://es.wikipedia.org/wiki/Wikipedia:Lista_de_art%C3%ADculos_que_toda_Wikipedia_deber%C3%ADa_tener'

response = requests.get(topics_url)
soup = BeautifulSoup(response.text, 'html.parser')
all_order_list = soup.find_all('ol')

es_all_topic_urls = []
for ol in all_order_list:
    for anchor in ol.find_all('a'):
        es_all_topic_urls.append(f"{base_url}{anchor['href']}")

print(f"Total No. of topics to search: {len(es_all_topic_urls)}")
```

```python
# Getting the article content for each link and writing content to file.

for topic in es_all_topic_urls:
    current_topic = topic[30:]

    wk.set_lang('es')
    try:
        page = wk.page(current_topic)
        tmp_page = list(filter(None, page.content.split('\n')))
        clean_page = '\n'.join(tmp_page)
    except Exception:
        print(f'TOPIC {current_topic} DOES NOT MATCH ANY PAGE.')

    with open(f'./es_data/es_{current_topic}.txt', 'w') as fo:
        fo.write(clean_page)
        print(f'Successful writing: es_{current_topic}.txt')
```
